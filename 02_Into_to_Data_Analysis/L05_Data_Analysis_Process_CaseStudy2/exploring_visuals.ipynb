# Exploring with Visuals
Use `clean_08.csv` and `clean_18.csv`

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
% matplotlib inline

# load datasets

df_08 = pd.read_csv('clean_08.csv')
df_18 = pd.read_csv('clean_18.csv')
```

```python
df_08.dtypes
```

```python
ax1 = df_08['greenhouse_gas_score'].hist();
ax1.set_xlabel('Greenhouse Gas Score')
ax1.set_ylabel('# of Occurences');
```

```python
ax2 = df_18['greenhouse_gas_score'].hist();
ax2.set_xlabel('Greenhouse Gas Score')
ax2.set_ylabel('# of Occurences');
```

```python
ax1 = df_08['cmb_mpg'].hist();
ax1.set_xlabel('Combined MPG')
ax1.set_ylabel('# of Occurences');
```

```python
ax2 = df_18['cmb_mpg'].hist();
ax2.set_xlabel('Combined MPG')
ax2.set_ylabel('# of Occurences');
```
