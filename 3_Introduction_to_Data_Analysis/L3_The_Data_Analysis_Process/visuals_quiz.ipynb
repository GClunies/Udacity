Use the space below to explore `powerplant_data_edited.csv` to answer the quiz questions below.

```python
# imports and load data
import pandas as pd
% matplotlib inline

df_power = pd.read_csv('powerplant_data_edited.csv')
df_power.info()
```

```python
# plot relationship between temperature and electrical output
df_power.plot(x='temperature', y ='energy', kind='scatter');
```

```python
# plot distribution of humidity
df_power['humidity'].plot(kind='hist');
```

```python
# plot box plots for each variable
df_power.plot(kind='box');
```
